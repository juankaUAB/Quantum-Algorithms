Notebook para hacer las primeras pruebas de circuitos básicos, ejecutar el simulador y mostrar los resultados...

```python
from qiskit import QuantumCircuit, assemble, Aer # assemble es la función para compilar el circuito QuantumCircuit, Aer es el simulador donde se ejecutará el circuito
from qiskit.visualization import plot_histogram # visualizar el resultado de la ejecución
```

```python
'''Circuito de prueba'''
circuito = QuantumCircuit(2) # objeto para crear el circuito
circuito.x(1) # añadimos puerta X al qubit 1
circuito.measure_all() # medimos los dos qubits (todos)
circuito.draw(output='mpl') # plot del circuito
```

```python
'''Creamos el simulador y ejecutamos el circuito creado'''
simulador = Aer.get_backend('aer_simulator')
qobj = assemble(circuito)
resultado = simulador.run(qobj).result()
histograma = resultado.get_counts()
plot_histogram(histograma)
```
